```python
import pandas as pd
df = pd.read_csv("/kaggle/input/weather-csv/weather.csv")
df
```

# creating df into pivot data

```python
df.pivot(index='date', columns = 'city')
```

```python
df.pivot(index='date', columns = 'city', values = 'temperature')
```

```python
df.pivot(index='date', columns = 'city', values = 'humidity')
```

```python
df.pivot(index='humidity', columns = 'city')
```

# PIVOT_TABLE

```python
df = pd.read_csv('/kaggle/input/weather2-csv/weather2.csv')
df
```

# pivot table will create average temp and humidity in city date wise

```python
df.pivot_table(index='date', columns = 'city')
```

```python
df.pivot_table(index='date', columns = 'city', margins=True)
```

```python
df.pivot_table(index='date', columns = 'city', aggfunc= 'mean')
```

```python
df.pivot_table(index='date', columns = 'city', aggfunc= 'diff')
```

```python
df.pivot_table(index='date', columns = 'city', aggfunc= 'count')
```
