# convert the string date in to datetime using "parse_dates

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/weather-data1-csv/weather_data.csv",parse_dates=['day'])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace=True)
df
```

# Fill every NaN fields with 0

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

# fill the null value with the value of previoud day or next day

```python
df
```

```python
new_df = df.fillna(method = 'ffill')
new_df
```

```python
new_df = df.fillna(method = 'bfill')
new_df
```

```python
new_df = df.fillna(method = 'bfill', axis = 'columns')
new_df
```

# limit 1 with ffill help to copy the value of a cell to next 1 cell down 

```python
new_df = df.fillna(method = 'ffill', limit =1)
new_df
```

# Interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method = 'time')
new_df
```

# Drop all the rows that has na value

```python
df
```

```python
new_df = df.dropna()
new_df
```

# drop only the rows where all are na

```python
new_df = df.dropna(how='all')
new_df
```

# thresh = N, this will keep only those rows that have at least N non-missing values

```python
new_df = df.dropna(thresh=2)
new_df
```
