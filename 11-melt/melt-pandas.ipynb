# MELT is used to transform and reshape data

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/weather-melt-csv/weather.csv")
df
```

```python
df1 = pd.melt(df, id_vars='day')
df1
```

```python
df1 = pd.melt(df, id_vars='day', value_vars='chicago')
df1
```

```python
df1 = pd.melt(df, id_vars='day', value_vars=['chicago', 'berlin'])
df1
```

# Filter the variable of chicago only

```python
df1 = pd.melt(df, id_vars='day', value_vars=['chicago', 'berlin'])
df1[df1['variable'] == 'chicago']
```

```python
df1 = pd.melt(df, id_vars = ['day'])
df1[df1['variable'] == 'chicago']
```

```python
df1 = pd.melt(df, id_vars='day', var_name = 'city', value_name = 'temp')
df1
```
