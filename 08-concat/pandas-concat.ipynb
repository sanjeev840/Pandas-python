# creating a dataframe

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ['mumbai', 'delhi', 'bangalore'],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ['new york', 'chicago', 'orlando'],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})

us_weather
```

# concat the 2 df

```python
df = pd.concat([india_weather, us_weather])
df
```

# to have the indexing in continuation, use ignore_index

```python
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

# to have an additional index

```python
df = pd.concat([india_weather, us_weather], keys=['india', 'us'])
df
```

```python
df.loc['us']
```

```python
temperature_df = pd.DataFrame({
    "city": ['mumbai', 'delhi', 'bangalore'],
    "temperature": [32, 45, 30]    
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ['mumbai', 'delhi', 'bangalore'],
    "windspeed": [7, 12, 9]    
})
windspeed_df
```

# Appending 2 df will add the rows in the result df

```python
df = pd.concat([temperature_df, windspeed_df])
df
```

# to make the df in same rows, axis =1 is used

```python
df = pd.concat([temperature_df, windspeed_df], axis =1)
df
```

# in case of data not sequencial, we can use "index" for concatination the df

```python
temperature_df = pd.DataFrame({
    "city": ['mumbai', 'delhi', 'bangalore'],
    "temperature": [32, 45, 30]    
}, index = [0, 1, 2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ['delhi', 'mumbai'],
    "windspeed": [7, 12]    
}, index = [1, 0])
windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
s = pd.Series(["Humid", "Dry", "Rain"], name = "event")
s
```

```python
df = pd.concat([temperature_df, s], axis = 1)
df
```
